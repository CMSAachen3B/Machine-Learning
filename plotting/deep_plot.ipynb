Interesting metrics to observe in any (classification) training are loss and accuracy. They can be extracted from any keras training using the `fit()` function. This function returns an object called `History`. `History.history` attribute is a record of training loss values and metrics values at successive epochs. These were dumped already in the training step to `.npy` files. 

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import json
import os
```

Load all metrics:

```python
# load acc and loss
loss = np.load('../results/loss.npy')
val_loss = np.load('../results/val_loss.npy')
acc = np.load('../results/acc.npy')
val_acc = np.load('../results/val_acc.npy')
```

Plot the trainings and validation loss as a function of the number of epochs. By that you can control and check for overtraining in your training.

```python
# plot loss
f = plt.figure()
plt.plot(loss)
plt.plot(val_loss)
plt.xlabel("epochs")
plt.ylabel("loss")
plt.legend(["training loss", "validation loss"], loc="best")
plt.show()
f.savefig("loss.png")
```

Same for accuracy...

```python
# plot accuracy
f = plt.figure()
plt.plot(acc)
plt.plot(val_acc)
plt.xlabel("epochs")
plt.ylabel("accuracy")
plt.legend(["training accuracy", "validation accuracy"], loc="best")
plt.show()
f.savefig("accuracy.png")
```
